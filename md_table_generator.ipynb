```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("catastrum.csv")
df.reset_index(drop=True)
```

```python
header = '|'
div = '|'
content = ''
col = df.columns

for head in col:
    header = header+f" {head} |"
    div = div+f" ------- |"
print(header)
print(div)

a = df.shape[0]
b = 0

while a > b:
    col = df.iloc[b].to_list()
    for at in col:
        content = content+f" {at} |"
    print(content)
    b += 1
    
table = header+div+content
```
